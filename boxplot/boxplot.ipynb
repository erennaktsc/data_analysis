```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('supermarket_sales.csv')
```

```python
data.head()
```

```python
sns.boxplot(data=data, x='Total' , width=0.5,
            showfliers=False,fliersize=1000,notch=True)
# fliersize=1000 aykırı değerlerin boyutunu ayarlar.
#  notch = True --> Medyanın istatistiksel güven aralığını göstermek için kullanılır.
# Örnek: Kutunun içinde medyanın bir çentiği olur, güven aralığını gösterir.
# showfliers=False aykırı değerler varsa gizler.
plt.title('Boxplot of Total')
plt.xlabel('Total')
plt.show()
```

```python
sns.boxplot(data=data, y='Total' , width=0.2,
            whis=10)
# whis → Bıyıkların Uzunluğunu Ayarlama
plt.title('Boxplot of Total')
plt.xlabel('Total')
plt.show()
```

```python
sns.boxplot(data=data , x='Payment' , y='Total',
            width=0.5)
plt.title('Boxplot of Total by Payment')
plt.xlabel('Payment')
plt.ylabel('Total')
plt.show()
```

```python
sns.boxplot(data=data , x='Payment' , y='Total',
            color='red',width=0.5,hue='Gender')
plt.title('Boxplot of Total by Payment')
plt.xlabel('Payment')
plt.ylabel('Total')
plt.show()
```

```python
sns.boxplot(data=data , x='Payment' , y='Total',
            color='red',width=0.5,hue='Gender',whis=1.5,
            showmeans=True,meanprops={'marker':'o',
                                      'markerfacecolor':'white',
                                     'markeredgecolor':'black',
                                     'markersize':10})
# whis → Bıyıkların Uzunluğunu Ayarlama
# showmeans=True parametresi mean değerlerini gösterir.
#meanprops, showmeans=True parametresi aktifken, grafikte ortalama değeri temsil eden marker'ın stilini belirler.

plt.title('Boxplot of Total by Payment')
plt.xlabel('Payment')
plt.ylabel('Total')
plt.show()
```

```python
sns.boxplot(data=data , x='Payment' , y='Total',
            color='red',width=0.5,hue='Gender',dodge=False,
            showmeans=True,meanprops={'marker':'o',
                                      'markerfacecolor':'white',
                                     'markeredgecolor':'black',
                                     'markersize':10})
# dodge=True/False → Hue Kategorilerini Yan Yana mı Üst Üste mi Çizeceğini Belirler.
# showmeans=True parametresi mean değerlerini gösterir.
plt.title('Boxplot of Total by Payment')
plt.xlabel('Payment')
plt.ylabel('Total')
plt.show()
```

```python
sns.boxplot(data=data , x='Payment' , y='Total',
            color='red',width=0.5,hue='Gender',
            showmeans=True,meanprops={'marker':'o',
                                      'markerfacecolor':'white',
                                     'markeredgecolor':'black',
                                     'markersize':8},
                                     palette='husl',linewidth=10)
# linewidth=10 parametresi çizgi kalınlığını değiştirir.
#palette='husl parametresi renk sıkalasını değiştirir.
# showmeans=True parametresi mean değerlerini gösterir.
plt.title('Boxplot of Total by Payment')
plt.xlabel('Payment')
plt.ylabel('Total')
plt.show()
```
