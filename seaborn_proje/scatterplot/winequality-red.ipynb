## Alkol ile şarap kalitesi arasındaki ilişkiyi gösteriniz.

```python
sns.scatterplot(data=df ,x='alcohol' , y='quality')
plt.title("Alkol ve Şarap Kalitesi")
plt.xlabel('Alkol')
plt.ylabel('Kalite')
plt.show()
```

## Yoğunluk (density) ile kalite ilişkisi, asiditeye göre renk ayrımıyla nasıl?

```python
sns.scatterplot(data=df ,x='density' , y='quality' , hue='fixed acidity',
                        palette='viridis')
plt.title("Yoğunluk, Kalite ve Sabit Asidite İlişkisi")
plt.xlabel("Yoğunluk")
plt.ylabel("Kalite")
plt.show()
```

## Sabit asit ile uçucu asit ilişkisi ve kaliteye göre büyüklük

```python
sns.scatterplot(data=df ,x='fixed acidity' , y='volatile acidity' ,
                            size="quality" , hue="quality", sizes=(20,200) , alpha=0.7,
                                legend="full")
plt.title("Sabit Asidite ve Uçucu Asidite (Kaliteye Göre Boyut)")
plt.xlabel("Sabit Asidite")
plt.ylabel("Uçucu Asidite")
plt.show()
```
