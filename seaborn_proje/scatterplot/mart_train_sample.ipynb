```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('mart_train_sample.cvs.xlsx').sample(50)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

##  Satış (Sales) ile Ürün Fiyatı (Item_MRP) arasındaki ilişki nedir?

```python
sns.scatterplot(data=df , x='Item_MRP', y='Sales')
plt.title('Ürün Fiyatı ve Satış İlişkisi')
plt.xlabel('Etiket Fiyatı')
plt.ylabel('Satış')
plt.show()
```

## Ürün türüne göre satış ve fiyat ilişkisi

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df ,x='Item_MRP' , y='Sales' , hue='Item_Type' ,
                                palette="Set2" ,alpha=0.8)
plt.show()
```

## Mağaza konum seviyesine göre satış ve fiyat ilişkisi

```python
sns.scatterplot(data=df ,x='Item_MRP' , y='Sales' , hue='Tier',
                            style='Tier' , palette="Set1" , s=60)
plt.title("Konum Seviyesine Göre Satış-Fiyat Dağılımı")
plt.xlabel("Etiket Fiyatı")
plt.ylabel("Satış")
plt.show()
```

## Ürün ağırlığı ile satış ilişkisi ve mağaza büyüklüğüne göre boyutlandırma

```python
size_dict = {'Small' : 40 , 'Medium' : 90 , 'High' : 140}
sns.scatterplot(data=df , x='Item_W' , y='Sales' , hue='Outlet_Size',
                        size='Outlet_Size' , sizes=size_dict , alpha=0.8)
plt.title("Mağaza Büyüklüğüne Göre Ağırlık-Satış İlişkisi")
plt.xlabel("Ürün Ağırlığı")
plt.ylabel("Satış")
plt.show()
```

## Ürün tipi ve mağaza açılış yılına göre satış-fiyat ilişkisi

```python
plt.figure(figsize=(15,9))
sns.scatterplot(data=df ,x='Item_MRP' , y='Sales' , hue='Item_Type',
                style='Outlet_Year' , palette="tab20", alpha=0.7 , legend="full")
plt.title("Ürün Tipi ve Mağaza Yılına Göre Satış-Fiyat")
plt.xlabel("Etiket Fiyatı")
plt.ylabel("Satış")
plt.show()
```

## Ürün tipi, mağaza seviyesi ve satış boyutu ile üçlü ayrım

```python
plt.figure(figsize=(15,9))
sns.scatterplot(data=df , x='Item_MRP' , y='Sales' , hue='Item_Type',
                            style='Tier' , size='Sales' , sizes=(40, 200),
                            palette="tab20" ,alpha=0.8)
plt.title("Ürün Tipi, Mağaza Seviyesi ve Satış Boyutlu Dağılım")
plt.xlabel("Etiket Fiyatı")
plt.ylabel("Satış")
plt.show()
```

## Mağaza açılış yılına göre satış ve fiyat ilişkisi

```python
sns.scatterplot(data=df , x='Item_MRP' , y='Sales' , hue='Outlet_Year' ,
                    palette="viridis" , marker="o" , alpha=0.7 , s=70)

plt.title("Mağaza Açılış Yılına Göre Satış-Fiyat")
plt.xlabel("Etiket Fiyatı")
plt.ylabel("Satış")
plt.show()
```
