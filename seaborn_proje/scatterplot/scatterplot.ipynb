```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

sns.scatterplot() â†’ Ä°ki sayÄ±sal deÄŸiÅŸken arasÄ±ndaki iliÅŸkiyi nokta grafiÄŸi (scatter plot) ile gÃ¶sterir.

Her nokta, bir veri gÃ¶zlemini temsil eder.

Ã–zellikle korelasyon veya sÄ±nÄ±f ayrÄ±mÄ± olup olmadÄ±ÄŸÄ±nÄ± anlamak iÃ§in kullanÄ±lÄ±r.

ðŸ“Œ KÄ±saca:

X ve Y ekseninde iki deÄŸiÅŸken seÃ§ilir â†’ veri noktalarÄ± Ã§izilir â†’ iliÅŸki olup olmadÄ±ÄŸÄ± gÃ¶rsel olarak analiz edilir.

```python
data = pd.read_excel('mart_train_sample.cvs.xlsx').sample(50)
data.head()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales')
plt.show()
```

```python
sns.scatterplot(data=data ,x='Item_MRP' , y='Sales' , s=200,
                color='purple',edgecolor='black',alpha=0.5)
# s=200 tÃ¼m noktalarÄ±n sabit boyutu
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales' , hue='Outlet_Size')
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales' , hue='Outlet_Size',
                style='Outlet_Size')
# style = noktalarÄ± ÅŸekillere gÃ¶re ayÄ±rÄ±r.
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales' , hue='Outlet_Size',
                style='Outlet_Size',markers={'High':"o",
                                             'Small':'*',
                                             'Medium':'v'})
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales',style='Outlet_Size',
                hue='Outlet_Size',
                palette='crest_r',
                markers={'High':"^",
                         "Small":"v",
                         "Medium":"o"})
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP' , y='Sales' , hue='Outlet_Size',
                size='Tier',
                sizes=(20,200),
                palette='Purples_r',
                markers={"High":"^","Small":"v","Medium":"o"})
# size= parametresiyle birlikte kullanÄ±lÄ±r.
# Ama boyutlarÄ±n hangi aralÄ±kta olacaÄŸÄ±nÄ± belirlemek iÃ§in sizes=(min, max) kullanÄ±rsÄ±n.
plt.show()
```

```python
sns.scatterplot(data=data , x='Item_MRP', y='Sales',hue='Outlet_Size',
                size='Tier',
                sizes=(20,200),
                legend='full',
                palette='Purples_r',
                markers={"High":"^" , "Small":"v" , "Medium":"o"})
# sizes=() -->Bir sÃ¼tundaki sayÄ±sal deÄŸerlere gÃ¶re her noktayÄ± kÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe otomatik Ã¶lÃ§ekler.
# AynÄ± anda hem hue= hem size= kullanabilirsin; bÃ¶ylece her nokta rengiyle bir kategoriyi, boyutuyla baÅŸka bir deÄŸeri gÃ¶sterir.
# legend=	AÃ§Ä±klama kutusu ayarÄ±	legend="brief" veya "full"
plt.show()
```

## Ã–rnek Ã‡alÄ±ÅŸma

```python
iris = sns.load_dataset("iris")

sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species",       # TÃ¼rlere gÃ¶re renk ayÄ±r
    style="species",     # TÃ¼rlere gÃ¶re ÅŸekil de ayÄ±r
    palette="Set1",      # Renk paleti
    s=70,                # Nokta boyutu
    alpha=0.8            # Hafif ÅŸeffaflÄ±k
)

plt.title("Iris Veri Seti - Petal UzunluÄŸu ve GeniÅŸliÄŸi Scatter Plot")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```
