```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df = df.rename(columns={
    "PassengerId": "YolcuNo",
    "Survived": "HayattaKaldı",
    "Pclass": "Sınıf",
    "Name": "İsim",
    "Sex": "Cinsiyet",
    "Age": "Yaş",
    "SibSp": "KardeşEş",
    "Parch": "EbeveynÇocuk",
    "Ticket": "BiletNo",
    "Fare": "Ücret",
    "Cabin": "Kamara",
    "Embarked": "BinişLimani"
})
```

```python
sns.boxplot(data=df  , x='Cinsiyet' , y='Yaş' ,
                    color='pink' , showmeans=True , width=0.6)
plt.title("Cinsiyete Göre Yaş Dağılımı")
plt.show()
```

## Sınıfa göre ücret dağılımı nasıldır?

```python
sns.boxplot(data=df , x='Sınıf' , y='Ücret' , color='gold',
                order=[1,2,3] , showmeans=True ,  notch=True)
plt.title('Sınıfa Göre Ücret Dağılımı')
plt.show()
```

## Hayatta kalma durumuna göre yaş dağılımı nedir?

```python
sns.boxplot(data=df ,x='HayattaKaldı' , y='Yaş' , hue='Cinsiyet',
                     showmeans=True , linewidth=2 )
plt.title('Hayatta Kalma  Durumuna Göre Yaş Dağılımı')
plt.show()
```

## Biniş Limanına göre ücret dağılımı nasıldır?

```python
sns.boxplot(data=df ,x='BinişLimani' , y='Ücret',
                        showfliers=False , color='skyblue')
plt.title("Biniş Limanına Göre Ücret Dağılımı")
plt.show()
```

## Hayatta kalma ve sınıfa göre ücret dağılımı ?

```python
sns.boxplot(data=df , x='Sınıf' , y='Ücret' , hue='HayattaKaldı',
                            showfliers=False, palette='Set3' , dodge=True, showmeans=True)
plt.title("Sınıf ve Hayatta Kalma Durumuna Göre Ücret Dağılımı")
plt.show()
```

```python
sns.boxplot(data=df , x='Sınıf' , y='Ücret' , hue='Cinsiyet',
                                order=[1,2,3], palette="Set2" , dodge=True,
                                notch=True , showmeans=True , showfliers=False,
                                width=0.7 , linewidth=1)
plt.title("Sınıf ve Cinsiyete Göre Ücret Dağılımı")
plt.xlabel("Sınıf")
plt.ylabel("Ücret")
plt.legend(title="Cinsiyet")
plt.show()
```
