```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('user_behavior_dataset.csv')
```

```python
df = df.rename(columns={
    "User ID": "KullanıcıID",
    "Device Model": "CihazModeli",
    "Operating System": "İşletimSistemi",
    "App Usage Time (min/day)": "UygulamaKullanımSüresi",
    "Screen On Time (hours/day)": "EkranAçıkSüresi_saat",
    "Battery Drain (mAh/day)": "PilTüketimi",
    "Number of Apps Installed": "YüklüUygulamaSayısı",
    "Data Usage (MB/day)": "VeriKullanımı",
    "Age": "Yaş",
    "Gender": "Cinsiyet",
    "User Behavior Class": "KullanıcıDavranışSınıfı"
})
```

## Cinsiyete göre uygulama kullanım süresi dağılımı nedir?

```python
sns.boxplot(data=df , x='Cinsiyet' , y='UygulamaKullanımSüresi' ,
                        color='skyblue' , showmeans=True  , notch=True  , width=0.6)
plt.title('Cinsiyete Göre Uygulama Kullanım Süresi(Dakika)')
plt.show()
```

##  Kullanıcı davranış sınıfına göre yaş dağılımı nasıldır?

```python
sns.boxplot(data=df ,x='KullanıcıDavranışSınıfı' , y='Yaş' 
                                ,color='purple' , showmeans=True ,showfliers=False , linewidth=2 )
plt.title('Kullanıcı Davranış Sınıfına Göre Yaş Dağılımı')
plt.show()
```

## İşletim sistemine göre pil tüketimi dağılımı nedir?

```python
sns.boxplot(data=df ,x='İşletimSistemi' , y='PilTüketimi'
                                ,showmeans=True , notch=True , color='orange')
plt.title("İşletim Sistemine Göre Pil Tüketimi")
plt.show()
```

## Cihaz modeline göre veri kullanımı (MB/gün) dağılımı nedir?

```python
sns.boxplot(data=df  , x='CihazModeli' , y='VeriKullanımı' , 
                        color='brown'  ,  showmeans=True , width=0.7 , linewidth=2)
plt.title('Cihaz Modeline Göre Veri Kullanımı (MB/gün)')
plt.xticks(rotation=45)
plt.show()
```

##   Cinsiyet ve kullanıcı davranış sınıfına göre uygulama kullanım süresi

```python
sns.boxplot(data=df , color='KullanıcıDavranışSınıf' , y='UygulamaKullanımSüresi'
                                ,hue='Cinsiyet' , palette='Set3' ,dodge=True , showmeans=True , width=0.6)
plt.title('Kullanıcı Davranış Sınıfı ve Cinsiyete Göre Kullanım Süresi')
plt.show()
```

## Yaş gruplarına göre pil tüketimi

```python
df['YaşGrubu'] = pd.cut(df['Yaş'] , bins=[0,18,30,45,60,120] , labels=["0-18","19-30","31-45","46-60","60+"])

sns.boxplot(data=df , x='PilTüketimi'  ,y='YaşGrubu'
                                     , notch=True  , showfliers=False , orient='h' , linewidth=2)
plt.title('Yaş Gruplarına Göre Pil Tüketimi')
plt.show()
```
