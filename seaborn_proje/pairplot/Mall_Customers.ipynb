```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df = df.rename(columns={
    "CustomerID": "MüşteriID",
    "Gender": "Cinsiyet",
    "Age": "Yaş",
    "Annual Income (k$)": "YıllıkGelir",
    "Spending Score (1-100)": "HarcamaSkoru"
})
```

## Cinsiyete göre müşterilerin yaş, yıllık gelir ve harcama skorunun dağılımı nasıldır?

```python
sns.pairplot(
    data=df,
    vars=['Yaş' , 'YıllıkGelir' , 'HarcamaSkoru'],
    hue='Cinsiyet',
    palette='Set2'
)
plt.suptitle('Cinsiyete Göre Özelliklerin Dağılımı', y=1.02)
plt.show()
```

## Harcama skoru ile yıllık gelir arasında kümelenmeler var mı?

```python
sns.pairplot(
    data=df,
    vars=['YıllıkGelir' , 'HarcamaSkoru'],
    hue='Cinsiyet',
    palette='coolwarm',
    kind='scatter',
    diag_kind='kde'
)
plt.suptitle("Yıllık Gelir & Harcama Skoru Arası Dağılım", y=1.02)
plt.show()
```

## Sadece kadın müşterilerde yaş ve harcama skoru ilişkisi nasıl?

```python
sns.pairplot(
    data=df[df['Cinsiyet']=="Female"],
    vars=['Yaş' , 'HarcamaSkoru'],
    kind='reg',
    plot_kws={'line_kws':{"color":"crimson"}},
    diag_kind='hist'
)
plt.suptitle("Kadınlarda Yaş ve Harcama Skoru İlişkisi", y=1.02)
plt.show()
```

## Tüm değişkenlerdeki dağılımları kategorik olarak cinsiyetle incele

```python
sns.pairplot(
    data=df,
    hue='Cinsiyet',
    palette='tab10',
    markers=['o' , 's'],
    plot_kws={"alpha":0.7},
    diag_kind="kde"
)
plt.suptitle("Tüm Değişkenlerde Cinsiyete Göre Dağılım", y=1.02)
plt.show()
```

## Yaş ile harcama skoru ilişkisini kümelere ayır 

```python
sns.pairplot(
    data=df,
    vars=['Yaş' , 'HarcamaSkoru'],
    hue='Cinsiyet',
    palette='husl',
    diag_kind='hist',
    height=3,
    aspect=1.2
)
plt.suptitle("Yaş & Harcama Skoru: Cinsiyete Göre" , y=1)
plt.show()
```
