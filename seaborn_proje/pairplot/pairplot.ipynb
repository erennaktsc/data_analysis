```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
data.shape
```

```python
data.head()
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
sns.pairplot(data=data , diag_kind='kde')
plt.show()
```

```python
sns.pairplot(data=data , kind='reg')
plt.show()
```

```python
sns.pairplot(data=data , hue='species')
plt.show()
```

```python
sns.pairplot(data=data , x_vars=['sepal_width' , 'petal_width' , 'petal_length'] ,
             y_vars=['sepal_length'] , diag_kind=None)
plt.show()
```

```python
sns.pairplot(data=data , corner=True)
plt.show()
```

```python
sns.pairplot(data=data , diag_kws=dict(color='green',kde=True))
plt.show()
```

```python
sns.pairplot(data=data , diag_kws=dict(color='green' , kde=True),
             plot_kws=dict(color='orange'))
plt.show()
```

```python
sns.pairplot(data=data , diag_kind=dict(color='green' , kde=True),
             plot_kws=dict(color='orange' , marker=10 , s=10))
plt.show()
```

```python
l = sns.pairplot(data=data , plot_kws=dict(color='purple'))
l.map_upper(sns.kdeplot)
l.map_lower(sns.kdeplot)
plt.show()
```

```python


# Profesyonel pairplot
sns.pairplot(
    data=data,
    hue="species",                # SÄ±nÄ±flarÄ± renklendir
    diag_kind="kde",              # Diagonal grafikleri KDE ile Ã§iz
    kind="scatter",               # Alt Ã¼Ã§gen scatter olsun
    palette="Set2",               # Renk paleti seÃ§imi
    corner=True,                  # GrafiÄŸin sadece alt Ã¼Ã§genini gÃ¶ster
    plot_kws={"s": 40, "alpha": 0.8, "edgecolor": "gray"},  # Nokta ayarlarÄ±
    diag_kws={"fill": True, "alpha": 0.6}                   # KDE Ã§izimi iÃ§in dolgu ve ÅŸeffaflÄ±k
)

plt.suptitle("Iris Veri Seti - Ã–zellikler ArasÄ± DaÄŸÄ±lÄ±m (Pairplot)", y=1.02, fontsize=14)
plt.show()
```
