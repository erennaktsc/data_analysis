## YaÅŸ ile oyuncu boyu/kilosu arasÄ±nda nasÄ±l bir iliÅŸki var?

```python
sns.pairplot(data=df
             ,vars=['age' , 'player_height' , 'player_weight']
             ,hue='draft_round'
             ,diag_kind='kde')
plt.suptitle('YaÅŸ-Boy-Kilo Ä°liÅŸkisi ve Draft Turu' , y=1.02)
plt.show()
```

## Ãœlkelere gÃ¶re oyuncu performans metrikleri nasÄ±l daÄŸÄ±lÄ±yor?

```python
df.head()
```

```python
plt.figure(figsize=(20,15))
sns.pairplot(df[df['country'].isin(df['country'].value_counts().index[:5])],
             vars=['pts','reb' , 'ast']
            ,hue='country'
            ,diag_kind='hist')
plt.suptitle('Ãœlkelere GÃ¶re Performans DaÄŸÄ±lÄ±mÄ±' , y=1.02)
plt.show()
```

## Draft yÄ±lÄ±na gÃ¶re oyuncu boyu ve kilosu nasÄ±l deÄŸiÅŸiyor?

```python
sns.pairplot(data=df
             ,vars=['player_height' , 'player_weight']
             ,hue='draft_year'
             ,palette='coolwarm')
plt.suptitle('Draft YÄ±lÄ±na GÃ¶re Boy-Kilo DeÄŸiÅŸimi' , y=1.02)
plt.show()
```

## TakÄ±m kÄ±saltmalarÄ±na gÃ¶re oyuncu boyu/kilosu ve yaÅŸ daÄŸÄ±lÄ±mÄ± nasÄ±l?

```python
df.head()
```

```python
top_teams = df['team_abbreviation'].value_counts().index[:5]

sns.pairplot(df[df['team_abbreviation'].isin(top_teams)]
             ,vars=['age' , 'player_height' , 'player_weight']
             ,hue='team_abbreviation')
plt.suptitle('TakÄ±mlara GÃ¶re Fiziksel Ã–zellikler')
```

## Oyun sÃ¼resi (gp) ile performans istatistikleri arasÄ±nda nasÄ±l bir korelasyon var?

```python
sns.pairplot(data=df
             ,vars=['gp' , 'pts' ,'reb' , 'ast']
             ,kind='reg'
             ,plot_kws={'line_kws':{'color':'red'}})
plt.suptitle('Oyun SayÄ±sÄ± ve Performans iliÅŸkisi' , y=1.02)
plt.show()
```

## Draft turuna gÃ¶re yaÅŸ ve deneyim (draft_year - age) nasÄ±l daÄŸÄ±lÄ±yor?

```python
df.head()
```

```python
df['experience'] = df['draft_year'] - (1996 - df['age'])

sns.pairplot(data=df
             ,vars=['age' , 'experience']
             ,hue='draft_round'
             ,diag_kind='hist')
plt.suptitle("Draft Turuna GÃ¶re YaÅŸ ve Deneyim", y=1.01)
plt.show()
```

```python
df['draft_year'] = pd.to_numeric(df['draft_year'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
df['experience'] = df['draft_year'] - (1996 - df['age'])

sns.pairplot(data=df
             ,vars=['age' , 'experience']
             ,hue='draft_round'
             ,diag_kind='hist')
plt.suptitle("Draft Turuna GÃ¶re YaÅŸ ve Deneyim", y=1.01)
plt.show()
```

## Pozisyonlara (boy/kilo oranÄ±) gÃ¶re performans metrikleri nasÄ±l deÄŸiÅŸiyor?

```python
df['size_ratio'] = df['player_height'] / df['player_weight']
sns.pairplot(data=df
             ,vars=['size_ratio' , 'pts' , 'reb' , 'ast']
             ,hue='draft_round'
             ,palette='muted')
plt.suptitle("Fiziksel Oran ve Performans Ä°liÅŸkisi" ,  y=1.05)
plt.show()
```
