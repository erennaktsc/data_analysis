## Yaş ile oyuncu boyu/kilosu arasında nasıl bir ilişki var?

```python
sns.pairplot(data=df
             ,vars=['age' , 'player_height' , 'player_weight']
             ,hue='draft_round'
             ,diag_kind='kde')
plt.suptitle('Yaş-Boy-Kilo İlişkisi ve Draft Turu' , y=1.02)
plt.show()
```

## Ülkelere göre oyuncu performans metrikleri nasıl dağılıyor?

```python
df.head()
```

```python
plt.figure(figsize=(20,15))
sns.pairplot(df[df['country'].isin(df['country'].value_counts().index[:5])],
             vars=['pts','reb' , 'ast']
            ,hue='country'
            ,diag_kind='hist')
plt.suptitle('Ülkelere Göre Performans Dağılımı' , y=1.02)
plt.show()
```

## Draft yılına göre oyuncu boyu ve kilosu nasıl değişiyor?

```python
sns.pairplot(data=df
             ,vars=['player_height' , 'player_weight']
             ,hue='draft_year'
             ,palette='coolwarm')
plt.suptitle('Draft Yılına Göre Boy-Kilo Değişimi' , y=1.02)
plt.show()
```

## Takım kısaltmalarına göre oyuncu boyu/kilosu ve yaş dağılımı nasıl?

```python
df.head()
```

```python
top_teams = df['team_abbreviation'].value_counts().index[:5]

sns.pairplot(df[df['team_abbreviation'].isin(top_teams)]
             ,vars=['age' , 'player_height' , 'player_weight']
             ,hue='team_abbreviation')
plt.suptitle('Takımlara Göre Fiziksel Özellikler')
```

## Oyun süresi (gp) ile performans istatistikleri arasında nasıl bir korelasyon var?

```python
sns.pairplot(data=df
             ,vars=['gp' , 'pts' ,'reb' , 'ast']
             ,kind='reg'
             ,plot_kws={'line_kws':{'color':'red'}})
plt.suptitle('Oyun Sayısı ve Performans ilişkisi' , y=1.02)
plt.show()
```

## Draft turuna göre yaş ve deneyim (draft_year - age) nasıl dağılıyor?

```python
df.head()
```

```python
df['experience'] = df['draft_year'] - (1996 - df['age'])

sns.pairplot(data=df
             ,vars=['age' , 'experience']
             ,hue='draft_round'
             ,diag_kind='hist')
plt.suptitle("Draft Turuna Göre Yaş ve Deneyim", y=1.01)
plt.show()
```

```python
df['draft_year'] = pd.to_numeric(df['draft_year'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
df['experience'] = df['draft_year'] - (1996 - df['age'])

sns.pairplot(data=df
             ,vars=['age' , 'experience']
             ,hue='draft_round'
             ,diag_kind='hist')
plt.suptitle("Draft Turuna Göre Yaş ve Deneyim", y=1.01)
plt.show()
```

## Pozisyonlara (boy/kilo oranı) göre performans metrikleri nasıl değişiyor?

```python
df['size_ratio'] = df['player_height'] / df['player_weight']
sns.pairplot(data=df
             ,vars=['size_ratio' , 'pts' , 'reb' , 'ast']
             ,hue='draft_round'
             ,palette='muted')
plt.suptitle("Fiziksel Oran ve Performans İlişkisi" ,  y=1.05)
plt.show()
```
