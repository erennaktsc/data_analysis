```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
data = pd.read_csv('supermarket_sales.csv')
data = data[['Gender' , 'Payment' , 'Unit price' , 'Quantity' , 'Total' , 'gross income']]
data.head(10)
```

```python
sns.kdeplot(data=data , x='Total')
plt.show()
```

```python
sns.kdeplot(data=data)
plt.show()
```

```python
sns.kdeplot(data=data , x='Total' , bw_adjust=0.5)
#bw_adjust	Band genilii (bandwidth) ayar; kkdaha keskin, bykdaha przsz
plt.show()
```

```python
sns.kdeplot(data=data , x='Total' , hue='Payment')
#multiple=  Birden fazla kategoriye ait KDE erilerinin izimini kontrol eder.
#zellikle hue= ile birlikte kullanlr.
plt.show()
```

```python
sns.kdeplot(data=data , x='Total' , log_scale=True)
# Veriyi logaritmik lekle gsterir.
# Normalde veriler dorusal (linear) olarak izilirken, log_scale=True dersen:
# X ekseni, Y ekseni veya her ikisi de logaritmik lee dnr.
# Bylece ok bykok kk deerleri ayn grafikte daha okunur hale getirirsin.
plt.show()
```

```python
sns.kdeplot(data=data , x='Total' , hue='Payment' , multiple='stack',
            linewidth=3 , palette='Dark2' , alpha=0.5)
plt.show()
```

```python
sns.kdeplot(data=data , x='Unit price' , y='gross income')
plt.show()
```

```python
sns.kdeplot(data=data , x='Unit price' , y='gross income' , hue='Gender')
plt.show()
```

```python
sns.kdeplot(data=data , x='Unit price' , y='gross income' , hue='Gender',
            fill=True)
# fill	Erinin altn doldurur	Younluu daha okunakl ve grsel hale getirmek iin
plt.show()
```

```python
sns.kdeplot(data=data , x='Unit price' , y='gross income' , hue='Gender',
            fill=True , levels=5)
# parametresi, iki boyutlu KDE (younluk) grafiklerinde ka tane younluk seviyesi (e younluk izgisi  contour level) gsterileceini kontrol eder.
# Ne Zaman Kullanlr?
# 2D KDE grafii izdiinde (yani hem x= hem y= verdiinde)
# Grafik iinde younluk konturlar (contour lines) grnr olur
# levels sayesinde bu konturlarn saysn kontrol edersin
plt.show()
```

```python
sns.kdeplot(data=data , x='Unit price' , y='gross income',
            hue='Gender' , fill=True , levels=5 , thresh=0.2)
# thresh, sns.kdeplot() fonksiyonunda dolgu (fill=True) akken kullanlr.
# Younluk deeri belirli bir eik deerinin (threshold) altnda ise, o alan doldurmaz.
# Bylece ok dk younluklu yerler grafikte grnmez ve sadece anlaml, youn blgeler vurgulanr.
plt.show()
```
