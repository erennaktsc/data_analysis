```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

## Sepal uzunluğunun genel dağılımı nasıldır?

```python
sns.kdeplot(data=df  ,x="SepalLengthCm", fill=True , bw_adjust=0.8)
plt.title('Sepal Uzunluğu Dağılımı')
plt.xlabel('Sepal Uzunluğu')
plt.show()
```

## Türlere göre sepal uzunluğu dağılımı nasıl değişiyor?

```python
sns.kdeplot(data=df ,x="SepalLengthCm" , 
            hue='Species' , fill=True, multiple="stack" , bw_adjust=0.9 , palette='Set2' )
plt.title('Türlere Göre Sepal Uzunluğu Dağılımı')
plt.xlabel('Sepal Uzunluğu')
plt.show()
```

## Sepal genişliği dağılımında her tür kendi içinde normalize edilsin

```python
sns.kdeplot(data=df , x="SepalWidthCm" , hue='Species',
            fill=True , bw_adjust=1 , common_norm=False , multiple="layer")
plt.title("Türlere Göre Sepal Genişliği Dağılımı")
plt.xlabel("Sepal Genişliği (cm)")
plt.show()
```

## Türlere göre taç yaprak uzunluğu ?

```python
sns.kdeplot(data=df ,x="PetalLengthCm" , hue='Species',
            cumulative=True , fill=True  , bw_adjust=1 , palette="Set1" )
plt.title("Türlere Göre Kümülatif Petal Uzunluğu ")
plt.xlabel("Petal Uzunluğu ")
plt.show()
```

## Sepal uzunluğu dağılımı sadece 4–7 cm arasında incelensin 

```python
sns.kdeplot(data=df ,x="SepalLengthCm" , fill=True,
                    clip=(4,7), bw_adjust=0.9 )
plt.title("Sepal Uzunluğu KDE (Sadece 4-7cm Arası)")
plt.xlabel("Sepal Uzunluğu (cm)")
plt.show()
```

## Sepal genişliği dağılımını logaritmik eksende incele

```python
sns.kdeplot(data=df ,  x="SepalWidthCm" , color="purple",
                    fill=True , log_scale=True , bw_adjust=1 )

plt.title("Sepal Genişliği KDE (Logaritmik Eksen)")
plt.xlabel("Sepal Genişliği (cm) - Log")
plt.show()
```

## Sepal uzunluğu ve petal uzunluğu ilişkisini 2 boyutlu KDE ile göster

```python
sns.kdeplot(data=df , x="SepalLengthCm" ,y="PetalLengthCm",
                    fill=True , cmap='viridis' , levels=7 , thresh=0.1)
plt.title("Sepal Uzunluğu & Petal Uzunluğu")
plt.xlabel("Sepal Uzunluğu")
plt.ylabel("Petal Uzunluğu")
plt.show()
```
