```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Cinsiyete göre bahşiş dağılımı nasıldır?

```python
df.head()
```

```python
sns.violinplot(data=df, x='sex', y='tip'
               ,color='skyblue')

plt.title('Cinsiyete Göre Bahşiş Dağılımı')
plt.show()
```

## Sigara içen ve içmeyen müşterilerin toplam hesap tutarları nasıl dağılım gösterir?

```python
df.head()
```

```python
sns.violinplot(data=df, x='smoker',y='total_bill'
               ,hue='sex'
               ,split=True
               ,palette='coolwarm')
plt.title('Sigara İçen ve İçmeyen Müşterilerin Toplam Hesap Tutarları')
plt.show()
```

## Günlere göre kişi başına düşen ödeme miktarları nasıl değişir?

```python
sns.violinplot(data=df, x='day', y='price_per_person'
               ,inner='quartile'
               ,density_norm='count'
               ,color='pink')

plt.title('Günlere Göre Kişi Başı Ödeme Dağılımı')
plt.show()
```

## Öğle ve akşam yemeklerinde bahşiş oranları nasıl farklılık gösterir?

```python
sns.violinplot(data=df, x='time', y='tip',
               hue='sex'
               ,palette='viridis'
               ,split=True
               ,inner='stick')

plt.title('Yemek Zamanına Göre Bahşiş Dağılımı')
plt.show()
```

## Grup büyüklüğüne göre toplam hesap tutarı nasıl değişir?

```python
df.head()
```

```python
sns.violinplot(data=df,x='size',y='total_bill'
               ,palette='rocket'
               ,bw_method=.2
               ,cut=0
               ,density_norm='width')

plt.title('Grup Büyüklüğüne Göre Toplam Hesap Dağılımı')
plt.show()
```

## Hafta sonu ve hafta içi bahşiş dağılımları nasıldır?

```python
df.head()
```

```python
df['weekend'] = df['day'].isin(['Sat' , 'Sun'])

sns.violinplot(data=df, x='weekend', y='tip'
               ,inner='box'
               ,color='pink')

plt.title('Hafta Somu - Hafta İÇi Bahşiş Dağılımı')
plt.show()
```

## Cinsiyet ve sigara içme durumuna göre bahşiş/toplam hesap oranı nasıldır?

```python
df.head()
```

```python
df['tip_pct'] = df['tip'] / df['total_bill']

sns.violinplot(data=df, x='sex', y='tip_pct' ,
               split=True
               ,inner='point'
               ,palette='dark')

plt.title('Cinsiyet Ve Sigara Durumuna Göre Bahşiş Oranı')
```

## Sigara içen ve içmeyenlerin günlere göre bahşiş dağılımı nasıldır?

```python
sns.violinplot(data=df, x='day' , y='tip'
               ,hue='smoker'
               ,palette='magma'
               ,inner='quart'
               ,split=True)

plt.show()
```
