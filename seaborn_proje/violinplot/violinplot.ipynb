```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('supermarket_sales.csv')
```

```python
data.head()
```

```python
sns.violinplot(data=data, y='Total')
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               palette='husl')
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               color='orange' , density_norm='width')
plt.show()

# density_norm='area'  Her violinin genişliğini neye göre ayarla? (area, count, width)
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True)
# split=True/False → Grupları Aynı Violin İçinde Gösterme
# hue ile belirlenen iki kategori (örn. Kadın & Erkek) için tek bir violin içinde iki taraflı gösterim yapar.
plt.show()
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile")
# inner="quartile"  Violin içinde 1. çeyrek, medyan ve 3. çeyrek çizgilerini gösterir.
# inner=None → İç kısımdaki çizgileri kaldırır.
# inner="box" → Kutunun içinde boxplot ekler.
# inner="stick" → Bireysel gözlemleri çubuklar halinde gösterir.
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile",
               bw_method=2)
# bw_method=2 → Violin genişliğini ayarlar.
plt.show()
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile",
              bw_method=2,cut=0.9)
# cut=0.9→ Violinlerin uçlarını keser.
plt.show()
```
