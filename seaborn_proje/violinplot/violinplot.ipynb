```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('supermarket_sales.csv')
```

```python
data.head()
```

```python
sns.violinplot(data=data, y='Total')
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               palette='husl')
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               color='orange' , density_norm='width')
plt.show()

# density_norm='area'  Her violinin geniliini neye gre ayarla? (area, count, width)
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True)
# split=True/False  Gruplar Ayn Violin inde Gsterme
# hue ile belirlenen iki kategori (rn. Kadn & Erkek) iin tek bir violin iinde iki tarafl gsterim yapar.
plt.show()
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile")
# inner="quartile"  Violin iinde 1. eyrek, medyan ve 3. eyrek izgilerini gsterir.
# inner=None   ksmdaki izgileri kaldrr.
# inner="box"  Kutunun iinde boxplot ekler.
# inner="stick"  Bireysel gzlemleri ubuklar halinde gsterir.
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile",
               bw_method=2)
# bw_method=2  Violin geniliini ayarlar.
plt.show()
```

```python
sns.violinplot(data=data,x='Payment',y='Total',
               hue='Gender',split=True, inner="quartile",
              bw_method=2,cut=0.9)
# cut=0.9 Violinlerin ularn keser.
plt.show()
```
