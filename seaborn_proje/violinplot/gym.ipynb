```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gym.csv')
```

```python
df.head()
```

```python
df = df.rename(columns={
    'Age': 'Yaş',
    'Gender': 'Cinsiyet',
    'Weight (kg)': 'Kilo (kg)',
    'Height (m)': 'Boy (m)',
    'Max_BPM': 'Maksimum Kalp Atış Hızı',
    'Avg_BPM': 'Ortalama Kalp Atış Hızı',
    'Resting_BPM': 'Dinlenme Kalp Atış Hızı',
    'Session_Duration (hours)': 'Egzersiz Süresi (saat)',
    'Calories_Burned': 'Yakılan Kalori',
    'Workout_Type': 'Egzersiz Türü',
    'Fat_Percentage': 'Yağ Oranı',
    'Water_Intake (liters)': 'Su Tüketimi (litre)',
    'Workout_Frequency (days/week)': 'Egzersiz Sıklığı (gün/hafta)',
    'Experience_Level': 'Deneyim Seviyesi',
    'BMI': 'Vücut Kitle İndeksi'
})
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

## Cinsiyete göre yaş dağılımı nasıl?

```python
df.head()
```

```python
sns.violinplot(data=df
               ,x='Cinsiyet'
               ,y='Yaş'
               ,inner='quartile'
               ,color='purple'
               ,alpha=0.5
               ,edgecolor='black'
               ,linewidth=1.2
               ,cut=0
               ,density_norm='width')

plt.title('Cinsiyete Göre Yaş Dağılımı')
plt.xlabel('Cinsiyet', fontsize=12)
plt.ylabel('Yaş' , fontsize=12)
plt.show()
```

## Egzersiz türüne göre yakılan kalori dağılımı nasıl?

```python
df.head()
```

```python
plt.figure(figsize=(12,8))

sns.violinplot(data=df
               ,x='Egzersiz Türü'
               ,y='Yakılan Kalori'
               ,width=0.9
               ,cut=0
               ,inner='stick'
               ,order=['Yoga' , 'HIIT' , 'Cardio' , 'Strength']
               ,color='orange')

plt.title('Egzersiz Türüne Göre Yakılan Kalori Dağılımı')
plt.xlabel('Egzersiz Türü')
plt.ylabel('Yakılan Kalori')
plt.show()
```

## Deneyim seviyesine göre maksimum kalp atış hızı dağılımı nasıl?

```python
df.head()
```

```python
sns.violinplot(data=df
               ,x='Deneyim Seviyesi'
               ,y='Maksimum Kalp Atış Hızı'
               ,hue='Cinsiyet'
               ,cut=0.1
               ,bw_method=0.2
               ,split=True)

plt.title('Deneyim Seviyesine Göre Max Kalp Atış Hızı Dağılımı' , fontsize=15)
```

## Cinsiyete göre vücut kitle indeksi (BMI) dağılımı nasıl?

```python
df.head()
```

```python
plt.figure(figsize=(8,6))

sns.violinplot(data=df
               ,x='Cinsiyet'
               ,y='Vücut Kitle İndeksi'
               ,hue='Cinsiyet'
               ,cut=0
               ,density_norm='width'
               ,split=True
               ,inner='quartile'
               )

plt.title('Cinsiyete Göre Vücut Kitle İndeksi Dağılımı')
plt.xlabel('Cinsiyet', fontsize=12)
plt.ylabel('Vücut Kitle İndeksi' , fontsize=12)
plt.show()
```

## Vücut kitle indeksi (BMI) aralıklarına göre kalori yakımı nasıl değişir?

```python
df.head()
```

```python
df['vki'] = pd.cut(df['Vücut Kitle İndeksi'] , bins=[0, 18.5, 25, 30, 100], labels=['Zayıf', 'Normal', 'Fazla Kilolu', 'Obez'])

plt.figure(figsize=(10,6))

sns.violinplot(data=df
               ,x='vki'
               ,y='Yakılan Kalori'
               ,hue='Cinsiyet'
               ,cut=0
               ,inner='box'
               ,order=['Zayıf' , 'Normal', 'Fazla Kilolu' , 'Obez']
               ,width=0.7)

plt.title("VKİ Kategorilerine Göre Kalori Yakımı", fontsize=14)
plt.show()
```

## Deneyim seviyesine ve cinsiyete göre su tüketimi dağılımı nasıl?

```python
df.head()
```

```python
sns.violinplot(data=df
               ,x='Deneyim Seviyesi'
               ,y='Su Tüketimi (litre)'
               ,hue='Cinsiyet'
               ,palette='Set3'
               ,split=True
               ,inner='quartile'
               ,cut=0
               ,density_norm='width')

plt.title('Deneyim Seviyesi ve Cinsiyete Göre Su Tüketimi Dağılımı')
plt.xlabel('Deneyim Seviyesi')
plt.ylabel('Su Tüketimi (litre)')
plt.show()
```

## Haftalık antrenman sıklığına göre su tüketimi nasıl değişir?

```python
df.head()
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(data=df
               ,x='Egzersiz Sıklığı (gün/hafta)'
               ,y='Su Tüketimi (litre)'
               ,hue='Cinsiyet'
               ,cut=0
               ,inner='quart'
               ,density_norm='width'
               ,palette='Set2'
               ,common_norm=False
               ,gridsize=50)

plt.title("Haftalık Antrenman Sıklığına Göre Su Tüketimi", fontsize=14)
plt.show()
```
