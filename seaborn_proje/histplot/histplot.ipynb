```python
data = pd.read_csv('supermarket_sales.csv')
data = data[['Gender' , 'Payment' , 'Unit price' , 'Quantity' , 'Total' , 'gross income']]
data.head(10)
```

```python
sns.histplot(data=data , x='Total')
plt.show()
```

```python
sns.histplot(data=data , y='Total')
plt.show()
```

```python
sns.histplot(data=data, x='Total' , binwidth=100)
# binwidth=100 parametresi, histogramdaki bir Ã§ubuÄŸun (bar'Ä±n) geniÅŸliÄŸini ayarlar.
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data= data , x='Total' , bins=np.arange(0 ,1100,50))
plt.xticks(np.arange(0 , 1100 ,50))
plt.show()
```

```python
sns.histplot(data=data , x='Total' , kde=True)
# kde=True â†’ YoÄŸunluk eÄŸrisini ekler.
plt.show()
```

```python
sns.histplot(data=data , x='Total' , hue='Payment' , multiple='stack')
# sns.histplot(multiple=â€¦) parametresi, bir hue deÄŸiÅŸkeniyle Ã§izilen birden fazla histogramÄ±n nasÄ±l yerleÅŸeceÄŸini kontrol eder. KullanabileceÄŸin baÅŸlÄ±ca seÃ§enekler ÅŸunlardÄ±r:
# "layer"	Her bir hue grubu grafiÄŸe Ã¼st Ã¼ste, saydamlÄ±k (alpha) ile birlikte Ã§izilir (varsayÄ±lan).
# "dodge"	Her bir hue grubu yan yana (dodge) Ã§ubuklar olarak gÃ¶sterilir; Ã§akÄ±ÅŸma olmaz.
# "stack"	Ã‡ubuklar Ã¼st Ã¼ste istiflenir, her kategori kendi renginde ama tek bir Ã§ubuk olarak birleÅŸir.
# "fill"	TÄ±pkÄ± "stack" gibi istifler, ama toplam yÃ¼ksekliÄŸi 1 olacak ÅŸekilde normalize eder (yÃ¼zdelik gÃ¶sterim).
# "proportion"	Her bir bar, o bardaki her bir hue grubunun payÄ±nÄ± gÃ¶sterir (stack deÄŸil; bar yÃ¼ksekliÄŸi yine frekans, ama iÃ§indeki renkler oransal).
plt.show()
```

```python
sns.histplot(data=data , x='Total' ,hue='Payment' , element='poly')
# element parametresi, sns.histplot()â€™un histogramÄ± nasÄ±l Ã§izeceÄŸini (hangi â€œgeometrik elemanâ€ ile) belirler. SeÃ§ilebilecek 
# Ã¼Ã§ deÄŸer var:
#"bars"	(VarsayÄ±lan) Klasik bar grafiÄŸi Ã§izer.
# "step"	Sadece Ã§ubuklarÄ±n kenar Ã§izgilerini (step curve) gÃ¶sterir; dolgu yok.
# "poly"	Dolgulu â€œpolygonâ€ ÅŸeklinde bir eÄŸri Ã§izer.
plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize=(12,4))

# Bars
plt.subplot(1,3,1)
sns.histplot(x="total_bill", data=tips, element="bars", bins=20)
plt.title("element='bars'")

# Step
plt.subplot(1,3,2)
sns.histplot(x="total_bill", data=tips, element="step", bins=20)
plt.title("element='step'")

# Poly
plt.subplot(1,3,3)
sns.histplot(x="total_bill", data=tips, element="poly", bins=20)
plt.title("element='poly'")

plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=data , x='Total', stat='probability')
# stat parametresi, histogramÄ±n yÃ¼ksekliÄŸini (y-ekseni deÄŸerini) ne olarak gÃ¶sterileceÄŸini belirler.
# Seabornâ€™da kullanabileceÄŸin baÅŸlÄ±ca seÃ§enekler ÅŸunlardÄ±r:
# "count"	(VarsayÄ±lan) Her binâ€™deki gÃ¶zlem sayÄ±sÄ±nÄ± (frekansÄ±) gÃ¶sterir.
# "frequency"	Her binâ€™deki gÃ¶zlem sayÄ±sÄ±nÄ±, bin geniÅŸliÄŸine bÃ¶lerek gÃ¶sterir (count/width).
# "density"	YoÄŸunluk gÃ¶sterir; tÃ¼m histogram alanÄ± toplamÄ± 1 olacak ÅŸekilde normalize eder.
# "probability"	Her binâ€™deki gÃ¶zlem sayÄ±sÄ±nÄ±n toplam gÃ¶zlem sayÄ±sÄ±na oranÄ±nÄ± gÃ¶sterir (relative frequency).
# "percent"	YÃ¼zde olarak gÃ¶sterir; probability Ã— 100.
plt.show()
```

```python
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=20,
    kde=True,
    multiple="stack",
    stat="density",
    element="bars",
    common_norm=False
)
# common_norm=False	hue gruplarÄ±nÄ±n ayrÄ± ayrÄ± normalize edilmesini saÄŸlar.Her sÄ±nÄ±fÄ±n kendi daÄŸÄ±lÄ±mÄ± anlaÅŸÄ±lÄ±r
plt.show()
```
