```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
pd.set_option('display.max_rows',None)
```

sns.histplot() Nedir?
histplot → Histogram grafiği çizer.
Yani sayısal verilerin dağılımını sütunlarla (bar'larla) gösterir.

🎯 Kısaca: Veride hangi aralıkta kaç tane değer olduğunu gösterir.

```python
data = pd.read_csv('supermarket_sales.csv')
data = data[['Gender' , 'Payment' , 'Unit price' , 'Quantity' , 'Total' , 'gross income']]
data.head(10)
```

```python
sns.histplot(data=data , x='Total')
plt.show()
```

```python
sns.histplot(data=data , y='Total')
plt.show()
```

```python
sns.histplot(data=data, x='Total' , binwidth=100)
# binwidth=100 parametresi, histogramdaki bir çubuğun (bar'ın) genişliğini ayarlar.
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data= data , x='Total' , bins=np.arange(0 ,1100,50))
plt.xticks(np.arange(0 , 1100 ,50))
plt.show()
```

```python
sns.histplot(data=data , x='Total' , kde=True)
# kde=True → Yoğunluk eğrisini ekler.
plt.show()
```

```python
sns.histplot(data=data , x='Total' , hue='Payment' , multiple='stack')
# sns.histplot(multiple=…) parametresi, bir hue değişkeniyle çizilen birden fazla histogramın nasıl yerleşeceğini kontrol eder. Kullanabileceğin başlıca seçenekler şunlardır:
# "layer"	Her bir hue grubu grafiğe üst üste, saydamlık (alpha) ile birlikte çizilir (varsayılan).
# "dodge"	Her bir hue grubu yan yana (dodge) çubuklar olarak gösterilir; çakışma olmaz.
# "stack"	Çubuklar üst üste istiflenir, her kategori kendi renginde ama tek bir çubuk olarak birleşir.
# "fill"	Tıpkı "stack" gibi istifler, ama toplam yüksekliği 1 olacak şekilde normalize eder (yüzdelik gösterim).
# "proportion"	Her bir bar, o bardaki her bir hue grubunun payını gösterir (stack değil; bar yüksekliği yine frekans, ama içindeki renkler oransal).
plt.show()
```

```python
sns.histplot(data=data , x='Total' ,hue='Payment' , element='poly')
# element parametresi, sns.histplot()’un histogramı nasıl çizeceğini (hangi “geometrik eleman” ile) belirler. Seçilebilecek 
# üç değer var:
#"bars"	(Varsayılan) Klasik bar grafiği çizer.
# "step"	Sadece çubukların kenar çizgilerini (step curve) gösterir; dolgu yok.
# "poly"	Dolgulu “polygon” şeklinde bir eğri çizer.
plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize=(12,4))

# Bars
plt.subplot(1,3,1)
sns.histplot(x="total_bill", data=tips, element="bars", bins=20)
plt.title("element='bars'")

# Step
plt.subplot(1,3,2)
sns.histplot(x="total_bill", data=tips, element="step", bins=20)
plt.title("element='step'")

# Poly
plt.subplot(1,3,3)
sns.histplot(x="total_bill", data=tips, element="poly", bins=20)
plt.title("element='poly'")

plt.tight_layout()
plt.show()
```

```python
sns.histplot(data=data , x='Payment',shrink=0.8)
# shrink --> barların genişliğini ayarladık
plt.show()
```

```python
sns.histplot(data=data , x='Total', stat='probability')
# stat parametresi, histogramın yüksekliğini (y-ekseni değerini) ne olarak gösterileceğini belirler.
# Seaborn’da kullanabileceğin başlıca seçenekler şunlardır:
# "count"	(Varsayılan) Her bin’deki gözlem sayısını (frekansı) gösterir.
# "frequency"	Her bin’deki gözlem sayısını, bin genişliğine bölerek gösterir (count/width).
# "density"	Yoğunluk gösterir; tüm histogram alanı toplamı 1 olacak şekilde normalize eder.
# "probability"	Her bin’deki gözlem sayısının toplam gözlem sayısına oranını gösterir (relative frequency).
# "percent"	Yüzde olarak gösterir; probability × 100.
plt.show()
```

```python
sns.histplot(
    data=tips,
    x="total_bill",
    hue="sex",
    bins=20,
    kde=True,
    multiple="stack",
    stat="density",
    element="bars",
    common_norm=False
)
# common_norm=False	hue gruplarının ayrı ayrı normalize edilmesini sağlar.Her sınıfın kendi dağılımı anlaşılır
plt.show()
```
