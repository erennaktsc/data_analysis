```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Students Social Media Addiction.csv')
```

```python
df.head()
```

```python
df = df.rename(columns={
    "Student_ID": "Ogrenci_ID",
    "Age": "Yas",
    "Gender": "Cinsiyet",
    "Academic_Level": "Akademik_Seviye",
    "Country": "Ulke",
    "Avg_Daily_Usage_Hours": "Ortalama_Gunluk_Kullanim_Saati",
    "Most_Used_Platform": "En_Cok_Kullanilan_Platform",
    "Affects_Academic_Performance": "Akademik_Performansi_Etkiliyor",
    "Sleep_Hours_Per_Night": "Gecelik_Uyku_Saati",
    "Mental_Health_Score": "Mental_Saglik_Skoru",
    "Relationship_Status": "Iliski_Durumu",
    "Conflicts_Over_Social_Media": "Sosyal_Medya_Catismasi_Sayisi",
    "Addicted_Score": "Bagimlilik_Skoru"
})
```

## Öğrencilerin günlük sosyal medya kullanım saatlerinin dağılımı nasıldır?

```python
sns.histplot(data=df , x='Ortalama_Gunluk_Kullanim_Saati' , 
             bins=15 , kde=True)
plt.title('Günlük Sosyal Medya Kullanım Saati Dağılımı')
plt.xlabel('Ortalama Günlük Kullanım Saati')
plt.ylabel('Öğrneci Sayısı')
plt.show()
```

## Kadın ve erkek öğrencilerin günlük sosyal medya kullanım saatleri farklı mı?

```python
sns.histplot(data=df , x='Ortalama_Gunluk_Kullanim_Saati' , hue='Cinsiyet',
             bins=15 ,kde=True , multiple='dodge')
plt.show()
```

## Akademik seviye arttıkça sosyal medya bağımlılık puanı (Bagimlilik_Skoru) nasıl değişiyor?

```python
sns.histplot(data=df , x='Bagimlilik_Skoru' ,
              hue='Akademik_Seviye' , multiple='stack' , bins=10)
plt.title('Akademik Seviyeye Göre Sosyal Medya Bağımlılık Puanı')
plt.xlabel('Bağamlılık Puanı')
plt.ylabel('Öğrenci Sayısı')
plt.show()
```

## En çok kullanılan sosyal medya platformuna göre günlük kullanım saatleri nasıl değişiyor?

```python
sns.histplot(data=df , x='Ortalama_Gunluk_Kullanim_Saati' ,
              hue='En_Cok_Kullanilan_Platform' , bins=15,multiple='stack')
plt.show()
```

## Uyku saatine göre bağımlılık puanı nasıl değişiyor?

```python
sns.histplot(data=df , x='Gecelik_Uyku_Saati' , hue='Bagimlilik_Skoru',
            multiple='dodge' , discrete=True)
plt.title('Uyku Saatine Göre Bağımlılık Puanı')
plt.xlabel('Gecelik Uyku Saati')
plt.ylabel('Öğrenci Sayısı')
plt.show()
```

## Mental sağlık skoru düşük olanlar ve yüksek olanların bağımlılık puanı dağılımı nasıl?

```python
df['Mental_Saglik_Seviyesi'] = pd.cut(df['Mental_Saglik_Skoru'], bins=[0,4,7,10],labels=['Düşük' , 'Orta' , 'Yüksek'])
sns.histplot(data=df, x='Bagimlilik_Skoru' 
             , hue='Mental_Saglik_Seviyesi',bins=10,multiple='stack')
plt.title('Mental Sağlık Seviyesine Göre Bağımlılık Puanı')
plt.xlabel('Bağımlılık Puanı')
plt.ylabel('Oran')
plt.show()
```

## Sosyal medya ile ilgili yaşanan çatışma sayısı dağılımı nasıldır?

```python
sns.histplot(data=df , x='Sosyal_Medya_Catismasi_Sayisi' ,
             bins=8 , kde=True)
plt.title('Sosyal Medya Kaynaklı Çatışma Sayısı Dağılımı')
plt.xlabel('Çatışma Sayısı')
plt.ylabel('Öğrenci Sayısı')
plt.show()
```
