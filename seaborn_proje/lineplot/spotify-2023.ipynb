```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify-2023.csv", encoding="latin1")
```

```python
df = df.rename(columns={
    "track_name": "ParÃ§a AdÄ±",
    "artist(s)_name": "SanatÃ§Ä±",
    "released_year": "YÄ±l",
    "released_month": "Ay",
    "released_day": "GÃ¼n",
    "in_spotify_playlists": "Spotify Ã‡alma Listesi",
    "in_spotify_charts": "Spotify Listesi",
    "streams": "Dinlenme",
    "bpm": "BPM",
    "key": "Ton",
    "mode": "Mod",
    "danceability_%": "Dans Edilebilirlik",
    "valence_%": "Olumluluk",
    "energy_%": "Enerji",
    "acousticness_%": "Akustiklik",
    "instrumentalness_%": "EnstrÃ¼mantal",
    "liveness_%": "CanlÄ±lÄ±k",
    "speechiness_%": "KonuÅŸmasallÄ±k",
    "explicit": "UyarÄ±",
    "track_genre": "TÃ¼r"
})
```

```python
df["Dinlenme"] = pd.to_numeric(df["Dinlenme"].astype(str).str.replace(",", ""), errors="coerce")
df["YÄ±l"] = pd.to_numeric(df["YÄ±l"], errors="coerce")
```

```python
df.columns = df.columns.str.strip()
```

```python
yillik = df.groupby("YÄ±l")["Dinlenme"].mean().reset_index()

sns.lineplot(data=yillik , x='YÄ±l' , y='Dinlenme' ,  
             linewidth=2.5 , marker="o" , color='green')
plt.title('YÄ±llara GÃ¶re Ortalama Dinlenme')
plt.xlabel('YÄ±l')
plt.ylabel('Ortlama Dinleme')
plt.show()
```

## YÄ±la ve TÃ¼r'e GÃ¶re Ortalama Dinlenme

```python
aylik = df.groupby("Ay")['Dans Edilebilirlik'].mean().reset_index()

sns.lineplot(data=aylik , x="Ay" , y="Dans Edilebilirlik",
            errorbar=('ci', 95) ,  marker="o" , linewidth=2.1 , color="purple")

plt.title("Aylara GÃ¶re Ortalama Dans Edilebilirlik")
plt.xlabel("Ay")
plt.ylabel("Dans Edilebilirlik (%)")
plt.show()
```

## Enerji ve YÄ±llara GÃ¶re Trend

```python
enerji_yil = df.groupby("YÄ±l")["Enerji"].mean().reset_index()

sns.lineplot(data=enerji_yil , x="YÄ±l" , y="Enerji" , 
                    linewidth=2.5 , marker="o" , color="orange")


plt.title("YÄ±llara GÃ¶re Ortalama Enerji")
plt.xlabel("YÄ±l")
plt.ylabel("Enerji (%)")
plt.show()
```

## Aylara GÃ¶re Ortalama Olumluluk

```python
ay_olumluluk = df.groupby("Ay")["Olumluluk"].mean().reset_index()
```

```python
sns.lineplot(data=ay_olumluluk , x="Ay" ,  y='Olumluluk' 
                            , linewidth = 2.2 , marker="o" ,  color = "purple" , errorbar=("ci" , 68))
plt.title("Aylara GÃ¶re Ortalama Olumluluk")
plt.xlabel("Ay")
plt.ylabel("Olumluluk")
plt.show()
```
