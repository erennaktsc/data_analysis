```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spotify-2023.csv", encoding="latin1")
```

```python
df.describe()
```

```python
df = df.rename(columns={
    "track_name": "Parça Adı",
    "artist(s)_name": "Sanatçı",
    "released_year": "Yıl",
    "released_month": "Ay",
    "released_day": "Gün",
    "in_spotify_playlists": "Spotify Çalma Listesi",
    "in_spotify_charts": "Spotify Listesi",
    "streams": "Dinlenme",
    "bpm": "BPM",
    "key": "Ton",
    "mode": "Mod",
    "danceability_%": "Dans Edilebilirlik",
    "valence_%": "Olumluluk",
    "energy_%": "Enerji",
    "acousticness_%": "Akustiklik",
    "instrumentalness_%": "Enstrümantal",
    "liveness_%": "Canlılık",
    "speechiness_%": "Konuşmasallık",
    "explicit": "Uyarı",
    "track_genre": "Tür"
})
```

```python
df["Dinlenme"] = pd.to_numeric(df["Dinlenme"].astype(str).str.replace(",", ""), errors="coerce")
df["Yıl"] = pd.to_numeric(df["Yıl"], errors="coerce")
```

```python
df.columns = df.columns.str.strip()
```

```python
yillik = df.groupby("Yıl")["Dinlenme"].mean().reset_index()

sns.lineplot(data=yillik , x='Yıl' , y='Dinlenme' ,  
             linewidth=2.5 , marker="o" , color='green')
plt.title('Yıllara Göre Ortalama Dinlenme')
plt.xlabel('Yıl')
plt.ylabel('Ortlama Dinleme')
plt.show()
```

## Yıla ve Tür'e Göre Ortalama Dinlenme

```python
aylik = df.groupby("Ay")['Dans Edilebilirlik'].mean().reset_index()

sns.lineplot(data=aylik , x="Ay" , y="Dans Edilebilirlik",
            errorbar=('ci', 95) ,  marker="o" , linewidth=2.1 , color="purple")

plt.title("Aylara Göre Ortalama Dans Edilebilirlik")
plt.xlabel("Ay")
plt.ylabel("Dans Edilebilirlik (%)")
plt.show()
```

## Enerji ve Yıllara Göre Trend

```python
enerji_yil = df.groupby("Yıl")["Enerji"].mean().reset_index()

sns.lineplot(data=enerji_yil , x="Yıl" , y="Enerji" , 
                    linewidth=2.5 , marker="o" , color="orange")


plt.title("Yıllara Göre Ortalama Enerji")
plt.xlabel("Yıl")
plt.ylabel("Enerji (%)")
plt.show()
```

## Aylara Göre Ortalama Olumluluk

```python
ay_olumluluk = df.groupby("Ay")["Olumluluk"].mean().reset_index()
```

```python
sns.lineplot(data=ay_olumluluk , x="Ay" ,  y='Olumluluk' 
                            , linewidth = 2.2 , marker="o" ,  color = "purple" , errorbar=("ci" , 68))
plt.title("Aylara Göre Ortalama Olumluluk")
plt.xlabel("Ay")
plt.ylabel("Olumluluk")
plt.show()
```

## Spotify Çalma Listesi Sayısına Göre Ortalama Enerji

```python
cl_enerji = df.groupby("Spotify Çalma Listesi")["Enerji"].mean().reset_index()
```

```python
sns.lineplot(data=cl_enerji , x='Spotify Çalma Listesi' , y="Enerji",
             linewidth=2.4  , marker="o" , color="blue" , alpha=0.7)
plt.title("Çalma Listesi Sayısına Göre Ortalama Enerji")
plt.xlabel("Çalma Listesi Sayısı")
plt.ylabel("Enerji (%)")
plt.show()
```
