```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('supermarket_sales.csv')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",capsize=0.9)
# Bu şekilde alfabedik sıralama yapar
plt.show()
```

```python
data["Product line"].value_counts("Product line").index
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index)
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'])
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
# hue_order= Bu parametre, renk tonu değişkeninin benzersiz değerlerinin grafikte görüneceği sırayı belirtmenize olanak tanır.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'],
            capsize=0.2)
#capsize=0.2 parametresi ile hata çubuklarının uçlarını belirleriz.
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
# hue_order= Bu parametre, renk tonu değişkeninin benzersiz değerlerinin grafikte görüneceği sırayı belirtmenize olanak tanır.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'],
            capsize=0.2,palette="cividis",ci=None,alpha=1)
#ci= None parametresi ile hata çubuklarını kaldırırız.  
#capsize=0.2 parametresi ile hata çubuklarının uçlarını belirleriz.
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
# hue_order= Bu parametre, renk tonu değişkeninin benzersiz değerlerinin grafikte görüneceği sırayı belirtmenize olanak tanır.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'],
            capsize=0.2,palette="cividis",ci=None,saturation=1)
# saturation=0.8 parametresi ile renk doygunluğunu belirleriz.
#ci= None parametresi ile hata çubuklarını kaldırırız.  
#capsize=0.2 parametresi ile hata çubuklarının uçlarını belirleriz.
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
# hue_order= Bu parametre, renk tonu değişkeninin benzersiz değerlerinin grafikte görüneceği sırayı belirtmenize olanak tanır.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'],
            capsize=0.2,palette="cividis",ci=None,saturation=1,
            estimator=sum)
# estimator=sum kullanıldığında, ortalama yerine toplam değerler (sum) alınarak çubuklar çizilir.
# saturation=0.8 parametresi ile renk doygunluğunu belirleriz.
#ci= None parametresi ile hata çubuklarını kaldırırız.  
#capsize=0.2 parametresi ile hata çubuklarının uçlarını belirleriz.
#order parametresi, grafikteki kategorik verilerin sıralamasını kontrol etmeni sağlar.
#  Varsayılan olarak, Seaborn verileri alfabetik veya frekans bazlı sıralayabilir, ancak order kullanarak hangi sırayla gösterileceğini kendin belirleyebilirsin.
# hue parametresi, grafiğe ek bir kategorik değişken ekleyerek veriyi alt gruplara ayırmanı sağlar.
# hue_order ise bu ek kategorinin sıralamasını belirler.
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Product line', y='Total',
            width=0.5,hue="Gender",order=data["Product line"].value_counts("Product line").index,
            hue_order=['Male' , 'Female'],
            capsize=0.2,palette="cividis",ci=None,saturation=1,
            estimator=np.median)
# estimator=median kullanıldığında, ortalama yerine medyan değerler alınarak çubuklar çizilir.
# saturation=0.8 parametresi ile renk doygunluğunu belirleriz.
#ci= None parametresi ile hata çubuklarını kaldırırız.  
#capsize=0.2 parametresi ile hata çubuklarının uçlarını belirleriz.
# order=... parametresi ile çubuklar en çok satılan (veya geçen) ürün hattından en az satılana doğru sıralanır.
# hue_order= Bu parametre, renk tonu değişkeninin benzersiz değerlerinin grafikte görüneceği sırayı belirtmenize olanak tanır.
plt.show()
```
