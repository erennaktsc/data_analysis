```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Hangi şube en yüksek toplam satışa sahiptir?

```python
df.head()
```

```python
sns.barplot(data=df ,x='Branch' , y='Total'
            ,hue='Branch'
            ,estimator=sum
            ,errorbar=None
            ,palette='viridis')
plt.title('Şubelere Göre Toplam Satış Miktarı')
plt.show()
```

## Hangi ürün kategorisinin ortalama birim fiyatı en yüksektir?

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df
            ,x='Product line'
            ,y='Unit price'
            ,hue='City'
            ,estimator=np.mean
            ,errorbar=('ci', 95))
plt.title('Ürün Kategorilerine Göre Ortalama Birim Fiyat')
plt.show()
```

## Hangi ödeme yöntemi (Cash, Ewallet, Credit card) en çok kullanılmıştır?

```python
df.head()
```

```python
sns.barplot(data=df
            ,x='Payment'
            ,y='Quantity'
            ,order=['Cash' , 'Ewallet' , 'Credit card']
            ,width=0.5
            ,errorbar=None
            ,color='orange'
            ,alpha=0.6
            ,estimator=sum)
plt.title('Ödeme Yöntemlerine Göre Satış Adedi')
plt.show()
```

## Kadın ve erkek müşterilerin üyelik durumlarına göre ortalama memnuniyet puanları nasıl değişir?

```python
df.head()
```

```python
sns.barplot(data=df, x='Gender' , y='Rating'
            ,hue='Customer type'
            ,palette='dark'
            ,alpha=0.8
            ,capsize=0.1
            ,err_kws={'linewidth': 1.5})
plt.title('Cinsiyet ve Üye Durumuna Göre Ortalama Rating')
plt.show()
```

## Hangi şehir en yüksek vergi gelirini sağlamıştır?

```python
df.head()
```

```python
sns.barplot(data=df
            ,x='City'
            ,y='Tax 5%'
            ,color='skyblue'
            ,errorbar=None
            ,width=0.5
            ,estimator=sum
            ,saturation=0.8)
```

## Günün hangi saatlerinde satışlar yoğunlaşmaktadır?

```python
df.head()
```

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
sns.barplot(data=df
            ,x='Hour'
            ,y='Total'
            ,hue='Branch'
            ,capsize=0.2
            ,estimator=sum)
plt.title('Günün Saatlerine Göre Satışlar')
plt.show()
```

##  Hangi ürün kategorisinin ortalama birim fiyatı en yüksektir?

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df ,x='Product line' ,y='Unit price'
            ,hue='City'
            ,palette='deep'
            ,estimator=np.mean
            ,errorbar=('ci', 95))
plt.title('Ürün Kategorilerine Göre Ortalama Birim Fiyat')
plt.show()
```
