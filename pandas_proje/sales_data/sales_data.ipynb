```python
df.rename(columns={
    'Date': 'Tarih',
    'Store ID': 'Magaza_ID',
    'Product ID': 'Urun_ID',
    'Category': 'Kategori',
    'Region': 'Bolge',
    'Inventory Level': 'Stok_Seviyesi',
    'Units Sold': 'Satilan_Adet',
    'Units Ordered': 'Siparis_Adedi',
    'Price': 'Fiyat',
    'Discount': 'Indirim',
    'Weather Condition': 'Hava_Durumu',
    'Promotion': 'Promosyon',
    'Competitor Pricing': 'Rakip_Fiyati',
    'Seasonality': 'Mevsimsellik',
    'Epidemic': 'Salgin',
    'Demand': 'Talep'
}, inplace=True)
```

## En yüksek ciroya sahip 10 satırı ve detaylarını göster.

```python
df['Ciro'] = df['Satilan_Adet'] * df['Fiyat']
df.nlargest(10, 'Ciro').reset_index()
```

## Bölge ve mağaza bazında toplam ciroyu sırala.

```python
result = df.groupby(['Bolge' , 'Magaza_ID'] , as_index=False)['Ciro'].sum()
result = result.sort_values('Ciro' , ascending=False)
result
```

## Her ürün için yıl bazında toplam satış adetini pivot tablo ile özetle.

```python
df['Yil'] = pd.to_datetime(df['Tarih']).dt.year
pivot = pd.pivot_table(df , values='Satilan_Adet' 
                       ,index='Urun_ID'
                       ,columns='Yil'
                       ,aggfunc='sum')
pivot
```

##  Fiyatı, kategorisinin ortalamasından %20 fazla olan ürünler ve detayları.

```python
kategori_ort = df.groupby('Kategori')['Fiyat'].transform('mean')
outlier = df[df['Fiyat'] > kategori_ort * 1.2]
outlier[['Urun_ID' , 'Kategori' , 'Fiyat']]
```

## Stok seviyesi ortalamanın altında ve satış adedi ortalamanın üstünde olan ürünler.

```python
stok_ort= df['Stok_Seviyesi'].mean()
satis_ort = df['Satilan_Adet'].mean()
filtre = df[(df['Stok_Seviyesi'] < stok_ort) & (df['Satilan_Adet']> satis_ort)]
filtre[['Urun_ID' , 'Stok_Seviyesi' , 'Satilan_Adet']]
```

##  En çok sipariş verilen 5 ürün ve toplam sipariş adedi.

```python
df.groupby('Urun_ID')['Siparis_Adedi'].sum().nlargest(5)
```

## Her mağaza için sezonluk ve sezonluk olmayan ürünlerin ortalama satış adedi.

```python
pivot = pd.pivot_table(data=df
                       ,values='Satilan_Adet'
                       ,index='Magaza_ID'
                       ,columns='Mevsimsellik'
                       ,aggfunc='mean')
pivot
```
