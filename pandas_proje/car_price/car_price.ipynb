## Yıla göre ortalama araba fiyatlarını bul ve artan sırada listele

```python
df.groupby('Year' , as_index=False)['Price'].mean().sort_values('Price')
```

## Hangi marka ve modelin, ortalama fiyatı en yüksek? İlk 5’i göster

```python
df.groupby(['Brand' , 'Model'])['Price'].mean().sort_values().head()
```

## Yakıt tipi (fuel) ve vites (transmission) kırılımında ortalama fiyat tablosu (pivot_table kullan)

```python
pivot = pd.pivot_table(df , values='Price' , index='Fuel_Type' ,
                        columns='Transmission' , aggfunc='mean')
pivot
```

##  Kilometresi 100.000’den az, fiyatı ortalamanın üstünde olan araçların sayısı

```python
df[(df['Mileage'] < 100_00)& (df['Price'] > df['Price'].mean())]
```

## Her markanın medyan fiyatı, toplam araç sayısı ve en düşük km’li aracı

```python
result = df.groupby('Brand').agg(
    medyan_fiyat = ('Price','median'),
    toplam_arac = ('Model','count'),
    min_km = ('Mileage','min')
)
result
```

## Fiyatı, markanın ortalamasının %30’undan fazla üzerinde olan araçları ve toplam sayısını getir

```python
mean_brand = df.groupby('Brand')['Price'].transform('mean')
outliers = df[df['Price'] > mean_brand *1.3]
outliers[['Brand' , 'Model' , 'Year' , 'Price']]
outliers.shape[0]
```
